# 1) **Descripción del Problema**
* <p><a href ="#ref7"> <font color=orange size=3> Inicio </font> </a></p>

En el contexto de un problema de riesgo por incumplimiento crediticio, se necesita un algoritmo para predecir si un futuro cliente caerá en default, y no pagará un prestamos solicitado. Nos encontramos ante un problema de clasificación binaria de aprendizaje supervisado. Ya que conocemos los valores que asume la variable target, 0 en caso de no default y 1 en caso de default

los entregables son; un análisis exploratorio de datos, el diseño, el procesamiento y el modelado de una solución en base a el aprendizaje automático, para predecir a que grupo pertenecerá el potencial cliente.

Se proporcionó un conjunto de datos para analizar y entrenar `application_train.parquet'` y otro con la descripción de cada campo que compone el conjunto de datos` application_metadata.parquet`

Para entrenar el predecir, se decidé trabajar por el metodo de Random Forest, por los siguientes motivos:<br>
* Proporciona muy buenos resultados.
* Se ejecuta de forma eficiente sobre grandes bases de datos.
* Puede manejar miles de variables sin tener que eliminar ninguna.
* Da estimaciones de la importancia de cada variable.
* Genera internamente un estimador insesgado del error de generalizació 
durante el proceso de construcción (OOB).
* Dispone de un m´etodo efectivo de estimaci´on de valores perdidos.
* Calcula proximidades entre pares de casos que pueden emplearse en el
análisis de conglomerados, identificación de “outliers”, o escalamiento de los datos para obtener representaciones gráficas.
* Si bien por detrás es un conjunto de Decision Trees, que son weak predictors. Cada uno de esos árboles es en si un mal predictor, están sobreentrenados sobre alguna parte específica del dataset. Pero Un conjunto de weak predictors con errores NO correlacionados (esto es muy importante) son capaces de entregar buenos resultados.El algoritmo Random Forest, consiste en entrenar una serie de árboles de decisión, para que luego voten el resultado por mayoría (la moda) en clasificación u obtengan la media si es un problema de regresión.
* No es necesario escalar variables numéricas. Si el algoritmo utiliza Gradient Descent (Logistic Regression, Linear Regression, Neural Networks por ejemplo) si es necesario.

## **Valores Nulos**

Los valores ausentes son muy importantes a la hora de crear modelos, algunos algoritmos no aceptan observaciones incompletas o bien se ven muy influenciados por ellas. Aunque la imputación de valores ausentes es parte del preprocesado y, por lo tanto, debe de aprenderse únicamente con los datos de entrenamiento, su identificación se tiene que realizar antes de separar los datos para asegurar que se establecen todas las estrategias de imputación necesarias. Para efectos de este ejercicio vamos a excluir los Valores ausentes o nulos con porcentajes mayor del 20%.

```python
nulos_porcentaje = (df.isna().sum())/len(df)*100
nulos_porcentaje.sort_values(ascending=False)
```
